```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

from kjm2601 import obsidian_media_path as path
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

# make a figure with 3 arrows illuatrating the 3D coordinate system

plt.xlim(-.2, 2)
plt.ylim(-.2, 2)
plt.axis('off')
arrow = patches.FancyArrowPatch((0,0), (2,0), arrowstyle='->', mutation_scale=20, linewidth=2)
ax.add_patch(arrow)
arrow = patches.FancyArrowPatch((0,0), (0,2), arrowstyle='->', mutation_scale=20, linewidth=2)
ax.add_patch(arrow)
arrow = patches.FancyArrowPatch((0,0), (1.2,1), arrowstyle='->', mutation_scale=20, linewidth=2)
ax.add_patch(arrow)
plt.text(2.1, 0, 'x', fontsize=20)
plt.text(0, 2.1, 'y', fontsize=20)
plt.text(1.3, 1.1, 'z', fontsize=20)

pos_e = [[.5, .7], [1.5, 1.5], [1.7, .28]]
pos_nuc = [[.21, 1.4], [.5, 1.6]]

for i,xy in enumerate(pos_e):
    x,y = xy
    arrow = patches.FancyArrowPatch((0,0), (x,y), arrowstyle='->', mutation_scale=20, linewidth=1, color='red')
    ax.add_patch(arrow)
    plt.plot(x, y, 'ro', markersize=10)
    plt.text(x+.05, y, f'$\mathbf{{r}}_{i+1}$', fontsize=12, horizontalalignment='left', verticalalignment='bottom')

for i,xy in enumerate(pos_nuc):
    x,y = xy
    arrow = patches.FancyArrowPatch((0,0), (x,y), arrowstyle='->', mutation_scale=20, linewidth=1, color='blue')
    ax.add_patch(arrow)
    plt.plot(x, y, 'bo', markersize=10)
    plt.text(x+.05, y, f'$\mathbf{{R}}_{i+1}$', fontsize=12, horizontalalignment='left', verticalalignment='bottom')

plt.title('Illustrasjon av elektronposisjoner og atomkjerner')
plt.savefig(path + 'elektronposisjoner_og_atomkjerner.svg')
plt.show()
```
