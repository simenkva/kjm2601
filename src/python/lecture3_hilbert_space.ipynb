```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib widget
from kjm2601 import obsidian_media_path as path
```

```python
# Define a wavefunction. This is a wavefunction for a moving particle modelled as a wave packet.

def make_psi(x, sigma = 2.0, k = 0.0, x0 = 0.0):
    return np.exp(-0.5*((x-x0)/sigma)**2)*np.exp(1j*k*x)
```

```python
def visualize_as_finite_vector(n = 10, labels=True):
    """ Visualize the wavefunction as a finite vector of length n
    by sampling at discrete points."""
    
    # Create a wavefunction
    x = np.linspace(-10, 10, 1000)
    psi = make_psi(x).real
    x1 = np.linspace(-10, 10, n)
    psi1 = make_psi(x1).real
    
    fig = plt.figure(figsize=(10, 5))
    plt.plot(x, psi, label='Wavefunction', linewidth=2, color='k')
    
    # increase font size for axes labels, ticks and title
    plt.xlabel('$x$', fontsize=16)
    plt.ylabel('$\psi(x)$', fontsize=16)
    
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    
    
    # plot vertical lines at the sample points
    for i in range(n):
        # add line from (x1[i], 0) to (x1[i], psi1[i]):
        plt.plot([x1[i], x1[i]], [0, psi1[i]], 'r--', linewidth=1)
        if labels:
            plt.text(x1[i] + .1, psi1[i] + .06, f'$y_{{{i}}}$', fontsize=12)
    plt.scatter(x1, psi1, color='r', label='Sample points')
    plt.xlabel('$x$')
    plt.ylabel('$\psi(x)$')
    if labels:
        plt.legend()
    filename = path + f'lecture3_wavefiunction_as_finite_vector_{n}.pdf'
    plt.title(f'Wavefunction as a finite vector, $n = {n}$', fontsize=20)
    plt.savefig(filename)
    plt.show()
```

```python
visualize_as_finite_vector(10)
visualize_as_finite_vector(20)
visualize_as_finite_vector(50, labels=False)
```
