```python
import numpy as np
import matplotlib.pyplot as plt
from kjm2601 import obsidian_media_path as path

# Parameters
x0 = 0.25 # Center of the Dirac delta function
epsilon = 0.0001  # Small width for approximation
x = np.linspace(-1, 1, 10000)  # x range

# Dirac delta function approximation (Gaussian)
delta_approx = (1 / (epsilon * np.sqrt(np.pi))) * np.exp(-((x - x0) ** 2) / (epsilon ** 2))

# Plot
plt.plot(x, delta_approx, label=r'$\delta_{x_0}(x)$')
plt.axvline(x=x0, color='r', linestyle='--', linewidth=0.5, label=r'$x_0$')
plt.title(f'Dirac Delta Function for $x_0 = {x0}$')
plt.xlabel('x')
plt.ylabel(r'$\delta_{x_0}(x)$')
plt.legend()
plt.grid(True)
plt.savefig(path + 'dirac_delta.pdf')
plt.show()
```
