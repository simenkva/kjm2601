```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib widget
from kjm2601 import obsidian_media_path as path
```

```python
# Define a wavefunction. This is a wavefunction for a moving particle modelled as a wave packet.

def make_psi(x, sigma = 1.0, k = 3.0, x0 = 0.0):
    return np.exp(-0.5*((x-x0)/sigma)**2)*np.exp(1j*k*x)

    
```

```python
# Create a wavefunction
x = np.linspace(-10, 10, 1000)
psi = make_psi(x)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, np.real(psi), label='Re $\psi$')
plt.plot(x, np.imag(psi), label='Im $\psi$')
plt.plot(x, np.abs(psi), color='k', linewidth=2, label='$|\psi|$')
plt.xlabel('x')
plt.ylabel('$\psi$')
plt.legend()
plt.title('Visualization of a wavefunction')
plt.savefig(path  + 'lecture3_wavefunction.pdf', bbox_inches='tight')
plt.show()

```
