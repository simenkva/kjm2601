```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
from kjm2601 import obsidian_media_path as path
```

```python
def linear_combo_of_plane_waves(x, k0, dk, N):
    if N > 1:
        k = np.linspace(-N/2, N/2, N+1)*dk + k0
    else:
        k = np.array([k0])
    y = np.zeros_like(x, dtype=complex)
    for i in range(N):
        y += np.exp(1j*k[i]*x)
    return y, k

```

```python
def plot_linear_combo_of_plane_waves(k0, dk, N, x_max):
    x = np.linspace(-x_max, x_max, 5000)
    y, k = linear_combo_of_plane_waves(x, k0, dk, N)
    plt.figure(figsize=(12, 6))
    plt.plot(x, np.real(y), label='real')
    plt.plot(x, np.imag(y), label='imag')
    plt.plot(x, np.abs(y), label='abs', color='black')
    if N > 1:
        plt.title(f'N={N}, |k| <= {np.max(k):.02f}, dk = {dk:.02f}')
    else:
        plt.title(f'N={N}, k0={k0:.2f}')
    plt.legend()
    filename = path + f'lecture6_linear_combo_of_plane_waves_N_{N}_k0_{k0:.2f}_dk_{dk:.2f}.pdf'
    print(f'saving to {filename}')
    plt.savefig(filename)
 
    plt.show()
```

```python
for N in [1, 3, 10, 100, 1000]:
    plot_linear_combo_of_plane_waves(k0 = 1, dk = 0.1, N=N, x_max=10)
```
