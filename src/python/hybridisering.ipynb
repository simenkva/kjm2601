```python
import matplotlib.pyplot as plt
import numpy as np
from kjm2601 import obsidian_media_path as path

# increase fint size for figures
plt.rcParams.update({'font.size': 14})


```

```python
R = 1.73

def H_1s(x, zeta = 1):
    return np.exp(-zeta * np.abs(x + R/2))

def F_2p(x, zeta = 1.5):
    return -np.exp(-zeta * np.abs(x - R/2)) * (x-R/2) * 2.5

x = np.linspace(-5, 5, 1000)
plt.figure()
chi1 = H_1s(x)
chi2 = F_2p(x)
plt.plot(x, chi1, label='1s, H')
plt.plot(x, chi2, label='2p$_z$ F')
plt.axvline(-R/2, linestyle='--')
plt.axvline(R/2, linestyle='--')
plt.xlabel('z')
plt.legend()
plt.title('AOs')
plt.savefig(path + '/HF_AOs.pdf')
plt.show()
```

```python
H = np.array([[-13.6, -8.35], [-8.35, -18.6]])
S = np.array([[1, 0.3], [0.3, 1]])

omega, C = np.linalg.eig(np.linalg.inv(S) @ H)

C_latex = np.round(C, 2)
omega_latex = np.round(omega, 2)
print(omega_latex)

print(C_latex)
latex_string = f'C = \\begin{{bmatrix}} {C_latex[0, 0]} & {C_latex[0, 1]} \\ {C_latex[1, 0]} & {C_latex[1, 1]} \\end{{bmatrix}}'
print(latex_string)

phi1 = C[0, 0] * chi1 + C[1, 0] * chi2
phi2 = C[0, 1] * chi1 + C[1, 1] * chi2

plt.figure()
plt.plot(x, phi1, label='$\phi_1$ (bonding)')
plt.plot(x, phi2, label='$\phi_2$ (antibonding)')
plt.axvline(-R/2, linestyle='--')
plt.axvline(R/2, linestyle='--')
plt.xlabel('z')
plt.legend()
plt.title('MOs')
plt.savefig(path + 'HF_MOs.pdf')
plt.show()
```

```python

plt.figure()
plt.plot(x, phi1**2, label='$\phi_1^2$')
plt.plot(x, phi2**2, label='$\phi_2^2$')
plt.axvline(-R/2, linestyle='--')
plt.axvline(R/2, linestyle='--')
plt.xlabel('z')
plt.legend()
plt.title('Charge density of MOs')
plt.savefig(path + 'HF_MOs_squared.pdf')
plt.show()
```

```python
c = C[:,0]
P_A = c[0]**2 + S[1, 0] * c[1] * c[0]
P_B = c[1]**2 + S[1, 0] * c[0] * c[1] 
print(P_A, P_B, P_A + P_B)
```
